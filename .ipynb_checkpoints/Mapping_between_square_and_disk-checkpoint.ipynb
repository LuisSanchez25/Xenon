This Notebook provides the functions necesay to the turn circles into squares and viceversa along with some illustrations of the transformations using Xenon data.

The paper used to create these fucntions can be found here: https://arxiv.org/ftp/arxiv/papers/1509/1509.06344.pdf

```python
import straxen
import numpy as np
import h5py
import matplotlib.pyplot as plt
import numpy as np
#import scipy.special.ellipkinc as ellip
#import scipy.special as scp
```

```python
#Access data
with h5py.File('/data/xenon1t/wfsim_data/wfsim_data_run_id_1001.h5', "r") as f:
    # List all groups
    print("Keys: %s" % f.keys())
    a_group_key = list(f.keys())[0]

    # This is the data you want to look
    data = np.array(f[a_group_key][:])
```

```python
#establish parameters for the construction of the lines in the scatter plot
epsilon = 0.02 #thickness of lines
num_data = np.size(data) #record number of data points
```

```python
#collect the locations of the data into two simple arrays for convinience
loc_x=[]
loc_y=[]
loop = np.arange(num_data)
for i in loop:
        loc_x = np.hstack((loc_x,(data['x'][i])))
        loc_y = np.hstack((loc_y,(data['y'][i])))
```

```python
#make range of the patterns between 0 and 1 for the sake of the formulas
u = np.zeros(int(num_data))
v = np.zeros(int(num_data))
count = np.arange(int(num_data))
for t in count:
    u[t] = loc_x[t]/(np.max(loc_x))
    v[t] = loc_y[t]/(np.max(loc_y))
```

```python
#First check of the data, all datapoints should be in a circle of radius 1
plt.scatter(u, v, marker='.', s=1)
```

```python
#Create different lines in the data to mark lines and see where they map to after transformations
line_marks_x = np.zeros(num_data)
line_marks_y = np.zeros(num_data)
f = 0 #placeholder variable
for c in count:
    #This first section generates the circles
    if np.sqrt(u[c]**2+v[c]**2) >= 0.2 - epsilon and np.sqrt(u[c]**2+v[c]**2) <= 0.2 + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    if np.sqrt(u[c]**2+v[c]**2) >= 0.5 - epsilon and np.sqrt(u[c]**2+v[c]**2) <= 0.5 + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    if np.sqrt(u[c]**2+v[c]**2) >= 0.7 - epsilon and np.sqrt(u[c]**2+v[c]**2) <= 0.7 + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    if np.sqrt(u[c]**2+v[c]**2) >= 0.9 - epsilon and np.sqrt(u[c]**2+v[c]**2) <= 0.9 + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    #this section generates lines through the circle
    if u[c] >= 0 - epsilon and u[c] <= 0 + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    if v[c] >= 0 - epsilon and v[c] <= 0 + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    if u[c] >= v[c] - epsilon and u[c] <= v[c] + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
    if u[c] >= -v[c] - epsilon and u[c] <= -v[c] + epsilon:
        line_marks_x[f] = u[c]
        line_marks_y[f] = v[c]
        f=f+1
```

```python
#This graph shows the results of the lines are circles created to mark down on the graphs
#these will serve as indicators of the change the functions have on the structure of the data
plt.scatter(u, v, marker='.', s=1, c='orange')
plt.scatter(line_marks_x, line_marks_y, marker='.', s=1, c ='red')
```

Now we can start experimenting with different transformations!

```python
#Simple streaching, similar code will be used for all other functions
x_s = np.zeros(int(num_data)) #new x,y coordinates
y_s = np.zeros(int(num_data))
x_lines_s = np.zeros(int(num_data)) #lines and circles generated on previous section
y_lines_s = np.zeros(int(num_data))
for n in count:
    #all the functions used can be found in the cited paper along with better illustrations
    if u[n]**2 >= v[n]**2:
        x_s[n] = np.sign(u[n])*np.sqrt(u[n]**2+v[n]**2)
        y_s[n] = np.sign(u[n])*(v[n]/u[n])*np.sqrt(u[n]**2+v[n]**2)
    if u[n]**2 < v[n]**2:
        x_s[n] = np.sign(v[n])*(u[n]/v[n])*np.sqrt(u[n]**2+v[n]**2)
        y_s[n] = np.sign(v[n])*np.sqrt(u[n]**2+v[n]**2)   
    if line_marks_x[n]**2 >= line_marks_y[n]**2:
        x_lines_s[n] = np.sign(line_marks_x[n])*np.sqrt(line_marks_x[n]**2+line_marks_y[n]**2)
        y_lines_s[n] = np.sign(line_marks_x[n])*(line_marks_y[n]/line_marks_x[n])*np.sqrt(line_marks_x[n]**2+line_marks_y[n]**2)
    if line_marks_x[n]**2 < line_marks_y[n]**2:
        x_lines_s[n] = np.sign(line_marks_y[n])*(line_marks_x[n]/line_marks_y[n])*np.sqrt(line_marks_x[n]**2+line_marks_y[n]**2)
        y_lines_s[n] = np.sign(line_marks_y[n])*np.sqrt(line_marks_x[n]**2+line_marks_y[n]**2)  
```

```python
#Scatter plot of Simple streaching, circles turned into squares
plt.scatter(x_s, y_s, marker='.', s=1, c='orange')
plt.scatter(x_lines_s, y_lines_s, marker='.', s=1, c ='red')
```

```python
#Now we create the inverse mapping which should hopefully reproduce back the original circle.
x_s_inv = np.zeros(int(num_data))
y_s_inv = np.zeros(int(num_data))
x_lines_s_inv = np.zeros(int(num_data))
y_lines_s_inv = np.zeros(int(num_data))
for w in count:
    if x_s[w]**2 >= y_s[w]**2:
        x_s_inv[w] = np.sign(x_s[w])*x_s[w]**2/np.sqrt(x_s[w]**2+y_s[w]**2)
        y_s_inv[w] = np.sign(x_s[w])*(y_s[w]*x_s[w])/np.sqrt(x_s[w]**2+y_s[w]**2)
    if x_s[w]**2 < y_s[w]**2:
        x_s_inv[w] = np.sign(y_s[w])*(x_s[w]*y_s[w])/np.sqrt(x_s[w]**2+y_s[w]**2)
        y_s_inv[w] = np.sign(y_s[w])*(y_s[w]**2)/np.sqrt(x_s[w]**2+y_s[w]**2)   
    if x_lines_s[w]**2 >= y_lines_s[w]**2:
        x_lines_s_inv[w] = np.sign(x_lines_s[w])*x_lines_s[w]**2/np.sqrt(x_lines_s[w]**2+y_lines_s[w]**2)
        y_lines_s_inv[w] = np.sign(x_lines_s[w])*(y_lines_s[w]*x_lines_s[w])/np.sqrt(x_lines_s[w]**2+y_lines_s[w]**2)
    if x_lines_s[w]**2 < y_lines_s[w]**2:
        x_lines_s_inv[w] = np.sign(y_lines_s[w])*(x_lines_s[w]*y_lines_s[w])/np.sqrt(x_lines_s[w]**2+y_lines_s[w]**2)
        y_lines_s_inv[w] = np.sign(y_lines_s[w])*y_lines_s[w]**2/np.sqrt(x_lines_s[w]**2+y_lines_s[w]**2)  
```

```python
plt.scatter(x_s_inv, y_s_inv, marker='.', s=1, c='orange')
plt.scatter(x_lines_s_inv, y_lines_s_inv, marker='.', s=1, c ='red')
#We get back the original circle! confirming the inverse
```

```python
#FG-Squircular Mapping
x_fg = np.zeros(int(num_data))
y_fg = np.zeros(int(num_data))
x_lines_fg = np.zeros(int(num_data))
y_lines_fg = np.zeros(int(num_data))
for i in count:
    x_fg[i] = (np.sign(u[i]*v[i])/(v[i]*np.sqrt(2)))*(np.sqrt(u[i]**2+v[i]**2-np.sqrt((u[i]**2+v[i]**2)*(u[i]**2+v[i]**2-4*(u[i]*v[i])**2))))
    y_fg[i] = (np.sign(u[i]*v[i])/(u[i]*np.sqrt(2)))*(np.sqrt(u[i]**2+v[i]**2-np.sqrt((u[i]**2+v[i]**2)*(u[i]**2+v[i]**2-4*(u[i]*v[i])**2))))
    x_lines_fg[i] = (np.sign(line_marks_x[i]*line_marks_y[i])/(line_marks_y[i]*np.sqrt(2)))*(np.sqrt(line_marks_x[i]**2+line_marks_y[i]**2-np.sqrt((line_marks_x[i]**2+line_marks_y[i]**2)*(line_marks_x[i]**2+line_marks_y[i]**2-4*(line_marks_x[i]*line_marks_y[i])**2))))
    y_lines_fg[i] = (np.sign(line_marks_x[i]*line_marks_y[i])/(line_marks_x[i]*np.sqrt(2)))*(np.sqrt(line_marks_x[i]**2+line_marks_y[i]**2-np.sqrt((line_marks_x[i]**2+line_marks_y[i]**2)*(line_marks_x[i]**2+line_marks_y[i]**2-4*(line_marks_x[i]*line_marks_y[i])**2))))
```

```python
plt.scatter(x_fg, y_fg, marker='.', s=1, c='orange')
plt.scatter(x_lines_fg, y_lines_fg, marker='.', s=1, c ='red')
```

```python
#Inverse FG-Squircular Mapping
x_fg_inv = np.zeros(int(num_data))
y_fg_inv = np.zeros(int(num_data))
x_lines_fg_inv = np.zeros(int(num_data))
y_lines_fg_inv = np.zeros(int(num_data))
for m in count:
    x_fg_inv[m] = x_fg[m]*np.sqrt(x_fg[m]**2+y_fg[m]**2-(x_fg[m]*y_fg[m])**2)/np.sqrt(x_fg[m]**2+y_fg[m]**2)
    y_fg_inv[m] = y_fg[m]*np.sqrt(x_fg[m]**2+y_fg[m]**2-(x_fg[m]*y_fg[m])**2)/np.sqrt(x_fg[m]**2+y_fg[m]**2)
    x_lines_fg_inv[m] = x_lines_fg[m]*np.sqrt(x_lines_fg[m]**2+y_lines_fg[m]**2-(x_lines_fg[m]*y_lines_fg[m])**2)/np.sqrt(x_lines_fg[m]**2+y_lines_fg[m]**2)
    y_lines_fg_inv[m] = y_lines_fg[m]*np.sqrt(x_lines_fg[m]**2+y_lines_fg[m]**2-(x_lines_fg[m]*y_lines_fg[m])**2)/np.sqrt(x_lines_fg[m]**2+y_lines_fg[m]**2)
```

```python
plt.scatter(x_fg_inv, y_fg_inv, marker='.', s=1, c='orange')
plt.scatter(x_lines_fg_inv, y_lines_fg_inv, marker='.', s=1, c ='red')
```

```python
#Elliptical Grid Mapping
x_e = np.zeros(int(num_data))
y_e = np.zeros(int(num_data))
x_lines_e = np.zeros(int(num_data))
y_lines_e = np.zeros(int(num_data))
for p in count:
    x_e[p] = (1/2)*np.sqrt(2+u[p]**2-v[p]**2+2*np.sqrt(2)*u[p])-(1/2)*np.sqrt(2+u[p]**2-v[p]**2-2*np.sqrt(2)*u[p])
    y_e[p] = (1/2)*np.sqrt(2-u[p]**2+v[p]**2+2*np.sqrt(2)*v[p])-(1/2)*np.sqrt(2-u[p]**2+v[p]**2-2*np.sqrt(2)*v[p])
    x_lines_e[p] = (1/2)*np.sqrt(2+line_marks_x[p]**2-line_marks_y[p]**2+2*np.sqrt(2)*line_marks_x[p])-(1/2)*np.sqrt(2+line_marks_x[p]**2-line_marks_y[p]**2-2*np.sqrt(2)*line_marks_x[p])
    y_lines_e[p] = (1/2)*np.sqrt(2-line_marks_x[p]**2+line_marks_y[p]**2+2*np.sqrt(2)*line_marks_y[p])-(1/2)*np.sqrt(2-line_marks_x[p]**2+line_marks_y[p]**2-2*np.sqrt(2)*line_marks_y[p])
```

```python
plt.scatter(x_e, y_e, marker='.', s=1, c='orange')
plt.scatter(x_lines_e, y_lines_e, marker='.', s=1, c ='red')
```

```python
x_e_inv = np.zeros(int(num_data))
y_e_inv = np.zeros(int(num_data))
x_lines_e_inv = np.zeros(int(num_data))
y_lines_e_inv = np.zeros(int(num_data))
for q in count:
    x_e_inv[q] = x_e[q]*np.sqrt(1-y_e[q]**2/2)
    y_e_inv[q] = y_e[q]*np.sqrt(1-x_e[q]**2/2)
    x_lines_e_inv[q] = x_lines_e[q]*np.sqrt(1-y_lines_e[q]**2/2)
    y_lines_e_inv[q] = y_lines_e[q]*np.sqrt(1-x_lines_e[q]**2/2)
```

```python
plt.scatter(x_e_inv, y_e_inv, marker='.', s=1, c='orange')
plt.scatter(x_lines_e_inv, y_lines_e_inv, marker='.', s=1, c ='red')
```
